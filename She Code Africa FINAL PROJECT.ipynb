```python
import pandas as pd
from sklearn import linear_model 
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
#Importing the dataframe used for analysis
df = pd.read_csv("../Adaaaa/Downloads/life-expectancy-who/Life Expectancy Data.csv")
df2=df.loc[df['Country']=='Nigeria']
df3 = df2[['Adult Mortality','GDP','percentage expenditure','Schooling']]

#This code test for linearity of the the variables adult mortality rate to the gdp of Nigerian economyfrom the year 2005 till 2015
#Linearity help to check if there is a relationship between the dependent and independent variable
a=df3['Adult Mortality']
y=df3['GDP']
plt.scatter(a,y,color ='green')
plt.title('GDP VS ADULT MORTALITY',fontsize=14)
plt.xlabel('Adult Mortality',fontsize=20)
plt.ylabel('GDP', fontsize=20)
plt.show()

#This code test for linearity of the the variables percentage expenditue rate to the gdp of Nigeria economyfrom the year 2005 till 2015

b=df3['percentage expenditure']
y=df3['GDP']
plt.scatter(b,y,color ='blue')
plt.title('GDP VS EXPENDITURE', fontsize=14)
plt.xlabel('expenditure',fontsize=20)
plt.ylabel('GDP', fontsize=20)
plt.show()

#This code test for linearity of the the variables schooling to the gdp of Nigeria economyfrom the year 2005 till 2015

c=df3['Schooling']
y=df3['GDP']
plt.scatter(c,y,color ='red')
plt.title('GDP VS SCHOOLING', fontsize=14)
plt.xlabel('Schooling',fontsize=20)
plt.ylabel('GDP', fontsize=20)
plt.show()
```

```python
#This are variables used for the multiple regression
X=df3[['Schooling','percentage expenditure','Adult Mortality']]
y=df3['GDP']
X=sm.add_constant(X)
result= sm.OLS(y,X).fit() 
result.summary()

##EXPLAING THE RESULTS OF THE MULTIPLE REGRESSION MODEL
#The dependent variable GDP is being predicted with independent variables such as schooling,percentage expenditure and adult mortality.
```
